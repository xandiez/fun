```python
import pandas as pd
import numpy as np


df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding = 'ISO=8859-1')

df
```

```python
#What is the total sum of Animal Fat produced in 2014 and 2017 respectively?
df1 = df.groupby ('Item') ['Y2014'] . sum()
print(df1)
```

```python
df12 = df.groupby ('Item') ['Y2017'] . sum()
print(df12)
```

```python
# What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?
df2 = df["Y2015"].mean()
print(round(df2,3))
df3 = df["Y2015"].std()
print(round(df3,3))
```

```python
# What is the total number and percentage of missing data in 2016 to 2 decimal places?
df4 = df.isnull()['Y2016'].sum()
print(df4)
df5= df.isnull()['Y2016'].count()
print(df5)
print(round((df4/df5)*100,2))
```

```python
# Which year had the highest correlation with ‘Element Code’?
corr2014 = df['Y2014'].corr(df['Element Code'])
print(corr2014)
corr2015 = df['Y2015'].corr(df['Element Code'])
print(corr2015)
corr2016 = df['Y2016'].corr(df['Element Code'])
print(corr2016)
corr2017 = df['Y2017'].corr(df['Element Code'])
print(corr2017)
corr2018 = df['Y2018'].corr(df['Element Code'])
print(corr2018)
```
